# Working with Pandas & Matplotlib

```python
# Filtering columns using Pandas & Showing the relationship with scatterplot

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a pandas DataFrame
df = pd.read_csv("/mxmh.csv")

# Filter the "Hours per day" and "Depression" columns
hours_depression_df = df[["Hours per day", "Depression"]]

# scatter plot visualization
plt.scatter(hours_depression_df["Depression"], hours_depression_df["Hours per day"])
plt.xlabel("Depression")
plt.ylabel("Hours per day")
plt.title("Depression vs Hours per day Scatter Plot")
```

Barplot w/ Pandas and matplotlib


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a pandas DataFrame
df = pd.read_csv("/mxmh.csv")

# Filter  "Hours per day" and "Depression" columns
hours_depression_df = df[["Hours per day", "Depression"]]

# Group data by Hours per day and calculate the mean depression score 
mean_depression_by_hours = hours_depression_df.groupby("Hours per day").mean()["Depression"]

# Create a bar plot to visualize the relationship between Hours per day and depression
ax = mean_depression_by_hours.plot(kind='bar', x=mean_depression_by_hours.index, y='Mean Depression Score', color='blue', legend=None)

# Set the x-axis and y-axis label
ax.set_xlabel('Hours per day')
ax.set_ylabel('Mean Depression Score')

# Set the plot title
ax.set_title('Relationship between Hours per day and Depression')

# Show  plot
plt.show()
```

# Seaborn Documentation Examples

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# a vlolinplot is A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.
# Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.
#not the best for small sample sizes
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex", kind="violin",
)
```

```python
sns.color_palette("cubehelix", as_cmap=True)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```
